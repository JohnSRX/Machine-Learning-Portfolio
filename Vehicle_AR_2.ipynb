```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv('/content/CARS.csv')
df.head()
```

```python
df=pd.read_csv('/content/CARS.csv', sep=',')
df.head()
```

```python
df=pd.read_csv('/content/CARS.csv', names=['Vehicles'])
df.head() 
```

```python
df.shape
```

```python
data=list(df["Vehicles"].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)

df=df.replace(False,0)
df=df.replace(True,1)

df
```

```python
df=apriori(df,min_support=0.2,use_colnames=True)
df
```

```python
df_ar=association_rules(df,metric="confidence", min_threshold=0.6)
df_ar
```

If you look at the first set of row, you can see that Ford and Toyota have a CONFIDENCE of .80, This is because there is an 80% chance that someone who owns a ford, also owns a toyota as their second vehicle. If you looka at the Tesla and Toyota row, you can see that they 67% chance of having both vehicles.

```python
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

If you look at the third row, you can see that Subaru and Honda have a LIFT of 2.285714, This is because there is a really positive coorilation between them. If you look at the ninth row, you can see that Tesla and Toyota have a LIFT of 1.025641. This means that they just have a slightly positive coorilation.

```python
df_ar=association_rules(df,metric="conviction", min_threshold=1.0)
df_ar
```
